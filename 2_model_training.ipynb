```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('Algerian_forest_fires_dataset_cleaned_dataset.csv')
```

```python
df.head()
```

```python
df.drop(['day','month','year'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
## Encoding
df['Classes']=np.where(df['Classes'].str.contains("not fire"),0,1)
```

```python
df['Classes'].value_counts()
```

```python
## divide our feature
## independent and dependent feature
X=df.drop('FWI',axis=1) # independent
y=df['FWI']
```

```python
X.head()
```

```python
y
```

```python
## Train,Test,Plate
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.shape,X_test.shape
```

```python
## Feature Selection based on correlation
X_train.corr()
```

```python
plt.figure(figsize=(12,10))
corr = X_train.corr()
sns.heatmap(corr,annot=True)
plt.show()
```

```python
## check for multicollinearity

def correlation(dataset,threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold: # to deal with -ve corr we take absolute value
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
## threshold
corr_features=correlation(X_train,0.85)
```

```python
## drop features when correlation is more than 0.85
X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis=1,inplace=True)
X_train.shape,X_test.shape
```

## feature scaling or standardization

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

## box plot to understand effect of  standard scaler

```python
plt.subplots(figsize=(15,5))
plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.title('X_train before scaling')
plt.subplot(1,2,2)
sns.boxplot(data=X_train_scaled)
plt.title('X_train after scaling')
plt.show()
```

```python
plt.subplots(figsize=(15,5))
plt.subplot(1,2,1)
sns.boxplot(data=X_test)
plt.title('X_test before scaling')
plt.subplot(1,2,2)
sns.boxplot(data=X_test_scaled)
plt.title('X_test after scaling')
plt.show()
```

## Linear Refression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
linreg=LinearRegression()
linreg.fit(X_train_scaled,y_train)
y_pred=linreg.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error:",mae)
print("R2_score:",score)
```

```python
y_test.shape,y_pred.shape
```

```python
## if the plot follows linear relationship the it is good model
plt.scatter(y_test,y_pred)
plt.show()
```

## Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
lasso=Lasso()
lasso.fit(X_train_scaled,y_train)
y_pred1=lasso.predict(X_test_scaled)
mae1=mean_absolute_error(y_test,y_pred1)
score1=r2_score(y_test,y_pred1)
print("Mean absolute error:",mae1)
print("R2_score:",score1)
```

```python
plt.scatter(y_test,y_pred1)
plt.show()
```

# Ridge regression

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
ridge=Ridge()
ridge.fit(X_train_scaled,y_train)
y_pred2=ridge.predict(X_test_scaled)
mae2=mean_absolute_error(y_test,y_pred2)
score2=r2_score(y_test,y_pred2)
print("Mean absolute error:",mae2)
print("R2_score:",score2)
```

```python
plt.scatter(y_test,y_pred2)
plt.show()
```
